```python
# –Ø—á–µ–π–∫–∞ 1: –ò–º–ø–æ—Ä—Ç –∏ –Ω–∞—Å—Ç—Ä–æ–π–∫–∞
import sys
import os
sys.path.append('..')

from ml_model import NeuralInvoiceParser
import matplotlib.pyplot as plt
import numpy as np

print("‚úÖ –ú–æ–¥—É–ª–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã")
```

```python
# –Ø—á–µ–π–∫–∞ 2: –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –ø–∞—Ä—Å–µ—Ä–∞
parser = NeuralInvoiceParser()
print("‚úÖ –ü–∞—Ä—Å–µ—Ä –≥–æ—Ç–æ–≤ –∫ —Ä–∞–±–æ—Ç–µ")
```

```python
# –Ø—á–µ–π–∫–∞ 3: –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
def visualize_invoice_results(result):
    """–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –ø–∞—Ä—Å–∏–Ω–≥–∞ —Å—á–µ—Ç–∞"""
    
    # –û—Å–Ω–æ–≤–Ω—ã–µ –ø–æ–ª—è –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
    main_fields = {
        'invoice_number': '–ù–æ–º–µ—Ä —Å—á–µ—Ç–∞',
        'date': '–î–∞—Ç–∞', 
        'total_amount': '–°—É–º–º–∞',
        'seller_inn': '–ò–ù–ù –ø—Ä–æ–¥–∞–≤—Ü–∞',
        'seller_name': '–ü—Ä–æ–¥–∞–≤–µ—Ü',
        'buyer_name': '–ü–æ–∫—É–ø–∞—Ç–µ–ª—å',
        'bik': '–ë–ò–ö',
        'bank_account': '–†–∞—Å—á–µ—Ç–Ω—ã–π —Å—á–µ—Ç',
        'kpp': '–ö–ü–ü',
        'phone': '–¢–µ–ª–µ—Ñ–æ–Ω',
        'email': 'Email',
        'company_address': '–ê–¥—Ä–µ—Å'
    }
    
    # –°–æ–∑–¥–∞–µ–º —Ñ–∏–≥—É—Ä—É
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
    
    # 1. –ì—Ä–∞—Ñ–∏–∫ –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω–æ—Å—Ç–∏ –ø–æ–ª–µ–π
    filled = []
    not_filled = []
    field_names = []
    
    for field, label in main_fields.items():
        field_names.append(label)
        if result.get(field):
            filled.append(1)
            not_filled.append(0)
        else:
            filled.append(0)
            not_filled.append(1)
    
    y_pos = np.arange(len(field_names))
    
    ax1.barh(y_pos, filled, color='#2ecc71', label='–ó–∞–ø–æ–ª–Ω–µ–Ω–æ')
    ax1.barh(y_pos, not_filled, left=filled, color='#e74c3c', label='–ù–µ –∑–∞–ø–æ–ª–Ω–µ–Ω–æ')
    ax1.set_yticks(y_pos)
    ax1.set_yticklabels(field_names)
    ax1.set_xlabel('–ó–∞–ø–æ–ª–Ω–µ–Ω–Ω–æ—Å—Ç—å')
    ax1.set_title('–ó–∞–ø–æ–ª–Ω–µ–Ω–Ω–æ—Å—Ç—å –ø–æ–ª–µ–π —Å—á–µ—Ç–∞')
    ax1.legend()
    
    # 2. –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –æ–±—Ä–∞–±–æ—Ç–∫–∏
    stats_text = [
        f"–í—Å–µ–≥–æ –ø–æ–ª–µ–π –Ω–∞–π–¥–µ–Ω–æ: {result.get('fields_found', 0)}",
        f"–°—Ç—Ä–∞–Ω–∏—Ü –≤ –¥–æ–∫—É–º–µ–Ω—Ç–µ: {result.get('total_pages', 0)}",
        f"–ö–∞—á–µ—Å—Ç–≤–æ —Ç–µ–∫—Å—Ç–∞: {result.get('text_quality', 'unknown')}",
        f"–ú–µ—Ç–æ–¥ –æ–±—Ä–∞–±–æ—Ç–∫–∏: {result.get('processing_method', 'unknown')}",
        f"ML –º–æ–¥–µ–ª—å: {'–≤–∫–ª—é—á–µ–Ω–∞' if result.get('ml_enabled') else '–≤—ã–∫–ª—é—á–µ–Ω–∞'}",
        f"–û–±–ª–∞—Å—Ç–µ–π –æ–±–Ω–∞—Ä—É–∂–µ–Ω–æ: {result.get('fields_detected', 0)}"
    ]
    
    ax2.text(0.1, 0.9, '\n'.join(stats_text), fontsize=12, 
             verticalalignment='top', linespacing=1.5)
    ax2.set_xlim(0, 1)
    ax2.set_ylim(0, 1)
    ax2.axis('off')
    ax2.set_title('–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –æ–±—Ä–∞–±–æ—Ç–∫–∏')
    
    plt.tight_layout()
    plt.show()
    
    # 3. –í—ã–≤–æ–¥ –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ–ª–µ–π
    print("\nüìã –ó–ù–ê–ß–ï–ù–ò–Ø –ü–û–õ–ï–ô:")
    print("=" * 50)
    for field, label in main_fields.items():
        value = result.get(field)
        if value:
            print(f"‚úÖ {label}: {value}")
    
    # 4. –û–±—â–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
    filled_count = sum(1 for field in main_fields if result.get(field))
    total_count = len(main_fields)
    print(f"\nüìä –ò–¢–û–ì–û: {filled_count}/{total_count} –ø–æ–ª–µ–π –∑–∞–ø–æ–ª–Ω–µ–Ω–æ ({filled_count/total_count*100:.1f}%)")

print("‚úÖ –§—É–Ω–∫—Ü–∏—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –≥–æ—Ç–æ–≤–∞")
```

```python
# –Ø—á–µ–π–∫–∞ 4: –ó–∞–ø—É—Å–∫ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏
def process_and_visualize(pdf_path):
    """–û–±—Ä–∞–±–æ—Ç–∞—Ç—å PDF –∏ –ø–æ–∫–∞–∑–∞—Ç—å –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—é"""
    
    print(f"üîç –û–±—Ä–∞–±–æ—Ç–∫–∞ —Ñ–∞–π–ª–∞: {pdf_path}")
    
    try:
        with open(pdf_path, 'rb') as f:
            pdf_bytes = f.read()
        
        result = parser.process_pdf(pdf_bytes)
        
        if 'error' in result:
            print(f"‚ùå –û—à–∏–±–∫–∞: {result['error']}")
            return
            
        # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—é
        visualize_invoice_results(result)
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –æ–±—Ä–∞–±–æ—Ç–∫–∏: {e}")

# –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º sample.pdf –µ—Å–ª–∏ –æ–Ω –µ—Å—Ç—å
if 'sample.pdf' in os.listdir('.'):
    process_and_visualize('sample.pdf')
else:
    print("‚ö†Ô∏è –§–∞–π–ª sample.pdf –Ω–µ –Ω–∞–π–¥–µ–Ω")
    print("–î–æ–±–∞–≤—å—Ç–µ PDF —Ñ–∞–π–ª –≤ –ø–∞–ø–∫—É notebooks –∏ –∏–∑–º–µ–Ω–∏—Ç–µ –∏–º—è —Ñ–∞–π–ª–∞ –≤ –∫–æ–¥–µ")
```
